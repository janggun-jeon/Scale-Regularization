```python
import os
import matplotlib.pyplot as plt

# amazon datasets log path
path = './log/rc-factor' 

# search for log files in the path
paths = []
for file_path in os.listdir(path):
    full_path = os.path.join(path, file_path)
    if os.path.isdir(full_path) and '.ipynb_checkpoints' != file_path:   
        paths.append(full_path)

# read data from log files
paths = sorted(paths)
info = []
for file in paths:
    data = []
    with open(file + '/gnn_auc.txt', 'r') as file:
        for line in file:
            data.append(float(line.strip()))
    info.append(data)
   
# initialize graph screen
plt.figure(figsize=(24, 16))
plt.boxplot(info, medianprops={'color': 'red'})
plt.xticks(range(1, 1 + len(paths)), [os.path.basename(seed).replace("_", "\n").replace("em", "embed") for seed in paths], fontsize=30, rotation=0)
plt.yticks(fontsize=25)  
plt.ylabel('AUROC', fontsize=18)
plt.legend()
plt.show()
```
